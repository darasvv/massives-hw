```python
import numpy as np
matrix = [
    {
        "A": [[1, 2, 4], [7, 5, -2], [-2, 1, 3]],
        "B": [[-1, 0, 3], [2, -6, 1], [-7, 1, 5]],
        "C": [[-2, 3], [4, 0], [7, -1]]
    },
    {
        "A": [[0, 3, 4], [-7, 1, -2], [2, -1, 3]],
        "B": [[1, 2, 4], [7, 0, -2], [-3, -2, 5]],
        "C": [[-3, 2], [0, 4], [-1, 7]]
    },
    {
        "A": [[5, 2, 9], [7, 0, -2], [-2, 1, 6]],
        "B": [[1, -2, 1], [7, 5, -2], [-2, 8, 3]],
        "C": [[-2, 3], [4, 5], [0, -1]]
    },
    {
        "A": [[1, -2, 8], [7, 5, -2], [-2, 3, 0]],
        "B": [[0, 2, 4], [1, -3, -2], [-2, 1, 3]],
        "C": [[0, 3], [5, -4], [7, -1]]
    },
    {
        "A": [[9, 2, 4], [7, 5, -5], [6, -2, 3]],
        "B": [[8, 2, 4], [0, 1, -3], [-2, 1, 6]],
        "C": [[-2, 0], [4, 8], [1, -1]]
    }
]

for idx, m in enumerate(matrix, start=1):
    A = np.array(m["A"])
    B = np.array(m["B"])
    C = np.array(m["C"])
    print("A+B:")
    print(A + B)
    print("B-A:")
    print(B - A)
    print("A*C:")
    print(A @ C)
    print("A*B*C:")
    print(A @ B @ C)
    print()
```
