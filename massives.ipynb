```python
class OneArray:
    def __init__(self, data: list = []):
        self.data = data
        self.size = len(data)
    def __str__(self):
        return f"{self.data}"
    def sum(self):
        return sum(self.data)
    def __add__(self, other):
        result = []
        for i in range(len(self.data)):
            sum_1 = self.data[i] + other.data[i]
            result.append(sum_1)
        return OneArray(result)

class TwoArray(OneArray):
    def __init__(self, row: int, col: int, data: list = []):
        super().__init__(data)
        self.row = len(data)
        self.col = len(data[0])
        self.matrix = col * row 
    def __str__(self):
        return "\n".join([" ".join(map(str, row)) for row in self.data])
    def sum(self):
        return sum(sum(row) for row in self.data)
    def __add__(self, other):
        result = []
        for row1, row2 in zip(self.data, other.data):
            new_row = [a + b for a, b in zip(row1, row2)]
            result.append(new_row)
        return TwoArray(result)
        
    def saddle_point(self):
            for i, row in enumerate(self.data):
                min_value = min(row)
                col_idx = row.index(min_value)
                column = [self.data[r][col_idx] for r in range(self.rows)]
                if min_value == max(column):
                    return i, col_idx, min_value
            return None
    def make_spiral(self):
        self.data = [[0] * self.cols for _ in range (self.rows)]
        directions = [(0, -1), (-1, 0), (0, 1), (1, 0)]
        d, num, x, y = 0, 1, self.cols - 1, self.rows - 1
        while num <= self.rows * self.cols:
            self.data[y][x] = num
            num += 1
            nx, ny = x + directions[d][1], y + directions[d][0]
            if nx<0 or nx>=self.cols or ny<0 or ny>=self.rows or self.data[ny][nx] !=0:
                d = (d + 1) % 4
                nx, ny = x + directions[d][1], y + directions[d][0]
                x, y = nx, ny

class ThreeArray(TwoArray):
    def __init__(self, data = None, depth = 0, rows = 0, cols = 0):
        self.rows = rows
        self.cols = cols
        self.depth = len(data)
        super().__init__(data)
    def __str__(self):
        result = []
        for i, layer in enumerate(self.data):
            result.append(f"Layer{i}:")
            for row in layer:
                result.append(" ".join(map(str, row)))
            result.append("")
        return "\n".join(result)
    def sum(self):
        total = 0
        for layer in self.data:
            for row in layer:
                for element in row:
                    total += element
        return total
    def __add__(self, other):
        result = []
        for lay1, lay2 in zip(self.data, other.data):
            new_layer = []
        for row1, row2 in zip(lay1, lay2):
            new_row = [a + b for a, b in zip(row1, row2)]
            new_layer.append(new_row)
        result.append(new_layer)
        return ThreeArray(result)
    def saddle_point_three(self):
        for d, layer in enumerate(self.data):
            for i, row in enumerate(layer):
                min_value = min(row)
                col_idx = row.index(min_value)
                column = [layer[r][col_idx] for r in range(len(layer))]
                if min_value == max(column):
                    return d, i, col_idx, min_value
            return None
```
